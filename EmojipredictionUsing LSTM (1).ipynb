```python
import numpy as np
import pandas as pd
import emoji

from keras.models import Sequential
from keras.layers import Dense, LSTM, SimpleRNN, Embedding

from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

from tensorflow.keras.utils import to_categorical
```

```python
data = pd.read_csv('/content/emoji_data.csv', header = None)
data.head()
```

```python
emoji_dict ={
    0: ":red_heart:",
    1: ":baseball:",
    2: ":grinning_face_with_big_eyes:",
    3: ":disappointed_face:",
    4: ":fork_and_knife_with_plate:"
}

def label_to_emoji(label):
  return emoji.emojize(emoji_dict[label])

```

```python
X = data[0].values
Y = data[1].values
X
```

```python
# embedding
file = open('/content/glove.6B.100d.txt','r',encoding = 'utf8')
content = file.readlines()
```

```python
embeddings = {}
for line in content:
  line = line.split()
  embeddings[line[0]] = np.array(line[1:], dtype = float)

print(len(embeddings) )
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(X)
word2index = tokenizer.word_index
word2index
```

```python
print(len(word2index))
```

```python
Xtokens = tokenizer.texts_to_sequences(X)
```

```python
# making all tokens of same length
def get_maxlen(data):
  maxlen = 0
  for sent in data:
    maxlen = max(maxlen, len(sent))
  return maxlen

maxlen = get_maxlen(Xtokens)



```

```python
Xtrain = pad_sequences(Xtokens, maxlen = maxlen, padding = 'post', truncating = 'post')

```

```python
Xtrain
```

```python
Ytrain = to_categorical(Y)
Ytrain
```

```python
embed_size = 100
embedding_matrix = np.zeros((len(word2index)+1, embed_size))

for word, i in word2index.items():
  # Check if word is present in embeddings before accessing
  if word in embeddings:
    embed_vector = embeddings[word]
    embedding_matrix[i] = embed_vector
  else:
    # Handle words not found in embeddings (e.g., assign a random vector)
    embedding_matrix[i] = np.random.rand(embed_size) # or any other strategy
```

```python
embedding_matrix
```

```python
#model creation

model = Sequential([
    Embedding(input_dim = len(word2index)+1,
              output_dim = embed_size,
              input_length = maxlen,
              weights = [embedding_matrix],
              trainable = False
              ),
    LSTM(units = 16, return_sequences= True), # return sequences return values at every time stamp
    LSTM(units = 4),
    Dense(5, activation ='softmax')
])

model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(Xtrain, Ytrain, epochs = 100)
```

```python
test = ["I feel good", "I feel very bad", "Lets eat dinner"]
test_seq = tokenizer.texts_to_sequences(test)
test_pad = pad_sequences(test_seq, maxlen = maxlen, padding = 'post', truncating = 'post')

y_pred = model.predict(test_pad)
y_pred = np.argmax(y_pred, axis = 1)

for i in range(len(test)):
  print(test[i], label_to_emoji(y_pred[i]) )
```
